```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/city_day.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
pollutants = ["PM2.5","PM10","NO2","SO2","CO","O3","NH3"] # officially used to compute AQI

for col in pollutants:
    df[col] = df[col].fillna(df[col].median())
```

```python
df = df.dropna(subset=["AQI"])
```

```python
df["Date"] = pd.to_datetime(df["Date"]) # convert to datetime
```

```python
top_cities = ["Delhi","Mumbai","Pune","Bengaluru","Kolkata"]
df = df[df["City"].isin(top_cities)]
```

```python
df = df.sort_values(["City","Date"])
```

```python
df.to_csv("../data/processed/aqi_clean.csv", index=False)
print("Cleaned dataset saved!")
print(df.shape)
```
